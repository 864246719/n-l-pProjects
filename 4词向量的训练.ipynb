```python
data_source = "sqlResult_1558435.csv"
```

```python
import pandas as pd
import jieba
```

```python
data = pd.read_csv(data_source,encoding="gb18030")
```

```python
def make_input_data(data):
    input_data = []
    for news in data:
        if type(news) is not str:
            continue
        for s in news.split():
            input_data.append(list(jieba.cut(s)))
    return input_data
```
